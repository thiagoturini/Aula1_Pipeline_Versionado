## Passo 2: Conceitos B√°sicos

### O que √© Git?

Git √© um **sistema de controle de vers√£o**.

Analogia: **M√°quina do Tempo para C√≥digo**
- Voc√™ pode tirar "fotos" do seu c√≥digo (commits)

### Termos Importantes:

- **Working Directory**: Seus arquivos atuais
- **Staging Area**: √Årea de prepara√ß√£o antes do commit

## Passo 7: .gitignore - Ignorar Arquivos

Nem tudo deve ser versionado!

### O que N√ÉO versionar:

1. **Segredos**: API keys, senhas
2. **Arquivos grandes**: Datasets gigantes
3. **Arquivos tempor√°rios**: Cache, logs
4. **Depend√™ncias**: Bibliotecas (podem ser reinstaladas)
5. **Arquivos gerados**: Compilados, processados

### Criar .gitignore:

````python
gitignore_content = """# Configura√ß√£o com segredos
config.json

# Ambiente virtual Python
venv/
.venv/
env/

# Cache Python
__pycache__/
*.pyc
*.pyo
.pytest_cache/

# Jupyter
.ipynb_checkpoints/

# IDEs
.vscode/
.idea/
*.swp

# Sistema
.DS_Store
Thumbs.db

# Dados (muito grandes)
dados/*.json
!dados/indice_filmes.json
"""

with open(".gitignore", "w") as f:
    f.write(gitignore_content)

print(".gitignore criado!")
````

## Passo 15: Conceito de Branches

Branches s√£o linhas do tempo paralelas.

### Analogia: Universos Paralelos

```
main:     A --- B --- C --- D
                      \
feature:              E --- F
```

- **main**: Branch principal (produ√ß√£o)
- **feature**: Branch de desenvolvimento

### Por que usar branches?

1. **Experimentos**: Teste sem medo
2. **Features**: Desenvolva funcionalidades isoladas
3. **Bugs**: Corrija sem afetar o principal
4. **Equipe**: Cada pessoa trabalha em seu branch

### Ver branches:

````python
!git branch
````

### Criar novo branch:

````python
!git branch feature/logging
````

````python
!git branch
````

### Mudar para o novo branch:

````python
!git checkout feature/logging
````

````python
!git branch
````
